```python
import pandas as pd
import numpy as np
from vnstock import Quote

#Date
start_date = "2015-01-01"
end_date = "2026-04-04"

#Download data
quote = Quote(symbol="VN30", source= "KBS")
df_vn30 = quote.history(start=start_date, end=end_date, interval="d")

df_vn30["time"] = pd.to_datetime(df_vn30["time"])
df_vn30.set_index("time", inplace=True)

df_vn30



```

#Feature engineering

```python
df_vn30["log_return"] = np.log(df_vn30["close"]/ df_vn30["close"].shift(1))

df_vn30["volatility"] = df_vn30["log_return"].rolling(window=14).std()

df_features = df_vn30.dropna(subset=["log_return", "volatility"]).copy()

df_features
```

```python
import os

# Create directory if it doesn't exist
os.makedirs("../data/processed", exist_ok=True)

# Save the DataFrame to a CSV file
df_features.to_csv("../data/processed/vn30_features.csv")
```
